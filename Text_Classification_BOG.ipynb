```python
from datasets import load_dataset

dataset = load_dataset("hossein20s/enrun-emails-text-classification")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def tokenize(input):
  bow = CountVectorizer(stop_words='english')
  bow.fit(input)
  b_f = bow.transform(input)
  output = b_f.toarray()
  return output
```

```python
inputs = dataset['train']['text']
```

```python
inputs = tokenize(inputs)
```

```python
targets = dataset['train']['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(inputs, targets, test_size=0.3, random_state=43)
```

```python
print(len(X_train))
print(len(Y_train))
print(len(X_test))
print(len(Y_test))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```
